```python
import os
import pandas as pd
import numpy as np
import shutil
```

```python
from tqdm.notebook import tqdm
```

```python
folder_path = "E:/User/bruna.fistarol/HMM/Salmonella/Profile"
```

```python
with open("genome_ids.txt", "r") as genomes:
    genome_ids = genomes.read().splitlines() 
```

```python
genome_ids
```

```python
profile = pd.DataFrame(index = genome_ids)

for genome in tqdm(genome_ids, desc='Processing genomes'):
    df_s = pd.read_csv(os.path.join(folder_path, "susceptibility_" + genome + ".tsv"), delimiter='\t')
    df_r = pd.read_csv(os.path.join(folder_path, "resistance_" + genome + ".tsv"), delimiter='\t')
    
    for antibiotic in df_s["genome_drug.antibiotic"]:
        profile.at[genome, antibiotic] = 0

    for antibiotic in df_r["genome_drug.antibiotic"]:
        profile.at[genome, antibiotic] = 1
```

```python
for col in profile.columns:
    print(f"\nFrequency count for {col}:")
    print(profile[col].value_counts(dropna=False))
```

```python
profile.to_csv('E:/User/bruna.fistarol/HMM/Salmonella/profile.csv')
```
